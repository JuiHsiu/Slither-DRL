```python
import matplotlib.pyplot as plt
import numpy
```

```python
def movingaverage(interval, window_size):
    window = numpy.ones(int(window_size))/float(window_size)
    return numpy.convolve(interval, window, 'valid')
```

# Policy Gradient

```python
draw_datas = []

with open('a2c_log.txt', 'r') as fin:
    tmp = fin.readlines()
    for value in tmp:
        draw_datas.append(float(value.strip()))
```

```python
plt.figure(figsize=(9, 6))
plt.title('Learning Curve of A2C')
plt.xlabel('steps (n=1000)')
plt.ylabel('avg. rewards')
plt.plot(draw_datas, label='A2C')
plt.legend(loc=4)
plt.show()
```
